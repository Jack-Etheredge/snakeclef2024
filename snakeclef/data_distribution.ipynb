```python
import pandas as pd

pd.set_option('display.max_rows', None)

from paths import METADATA_DIR

folder_metas = [("train", "SnakeCLEF2023-TrainMetadata-iNat.csv"),
                ("train", "SnakeCLEF2023-TrainMetadata-HM.csv"),
                ("val", "SnakeCLEF2023-ValMetadata.csv")]

train_df = pd.concat([pd.read_csv(METADATA_DIR / "SnakeCLEF2023-TrainMetadata-iNat.csv"),
                     pd.read_csv(METADATA_DIR / "SnakeCLEF2023-TrainMetadata-iNat.csv")], ignore_index=True)
val_df = pd.read_csv(METADATA_DIR / "SnakeCLEF2023-ValMetadata.csv")
```

```python
train_df.drop_duplicates("observation_id", inplace=True)
val_df.drop_duplicates("observation_id", inplace=True)
```

```python
train_counts = train_df.class_id.value_counts().reset_index().rename(columns={"count": "train_count"})
```

```python
val_counts = val_df.class_id.value_counts().reset_index().rename(columns={"count": "val_count"})
```

```python
combined_counts = train_counts.merge(val_counts, on="class_id", how="outer").fillna(0).astype(int)
combined_counts["val_percent"] = combined_counts["val_count"] / combined_counts["train_count"]
combined_counts["total"] = combined_counts["val_count"] + combined_counts["train_count"]
```

```python
combined_counts.sort_values(["total", "val_percent"])
```
